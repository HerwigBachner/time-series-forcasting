### Air Passengers Example Data Preparation

This dataset provides monthly totals of a US airline passengers from 1949 to 1960.

This notebook shows some charts about the content of the dataset.
To see how the forecasting solutions of the examples performs on real data, I split the data in a training set which contains the time series till the end of 1959 and a 'real data' series starting from 1960/01 till 1960/12.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df=pd.read_csv("../data/AirPassengers.csv")
df.columns=["Period","Passengers"]
df.info()
```

```python
df
```

```python
plt.figure(figsize=(20,10))
plt.plot(df["Passengers"])
plt.xlabel("Period")
labels=[df["Period"].iloc[x] for x in range(0,df.shape[0],12)]
plt.xticks(range(0,df.shape[0],12),labels=labels, rotation=45)
plt.ylabel("Number of Passenger")
plt.title("US Airline Passenger from 1949 to 1960")
plt.grid(True)
plt.show()
```

### Split the dataset
For testing purposes, I split the dataset in a training set (AirPassengersTraining.csv) and a comparison set which only contains the data from 1960/01 till 1960/12 (AirPassengers1960.csv)

```python
df.iloc[-12:].to_csv("../data/AirPassengersTraining.csv")
df.iloc[:-12].to_csv("../data/AirPassengers1960.csv")
```
